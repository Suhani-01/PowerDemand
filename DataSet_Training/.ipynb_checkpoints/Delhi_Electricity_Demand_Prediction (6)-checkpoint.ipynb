## 🔧 5. Data Preprocessing & Feature Engineering

```python
# ── 5.1 Merge SLDC + Weather on datetime ──
wx_feats = wx[['datetime','Temperature (°C)','Relative Humidity (%)',
               'Apparent Temperature (°C)','Precipitation (mm)',
               'Wind Speed (m/s)','Cloud Cover Total (%)']].copy()
wx_feats.rename(columns={'Apparent Temperature (°C)':'feels_like'}, inplace=True)

df = pd.merge(sldc, wx_feats, on='datetime', how='inner')
print(f'Merged shape: {df.shape}')

# ── 5.2 Remove Outliers (Winsorize at 0.5–99.5%) ──
print('\nRemoving outliers...')
before = len(df)
for r in REGIONS:
    lo, hi = df[r].quantile(0.005), df[r].quantile(0.995)
    df[r] = df[r].clip(lo, hi)

# Remove the BRPL spike explicitly
df = df[df['BRPL'] < 10000].copy()
after = len(df)
print(f'Outliers handled: {before - after} rows removed/clipped')

# ── 5.3 Time Features ──
df['hour']        = df['datetime'].dt.hour
df['day_of_week'] = df['datetime'].dt.dayofweek
df['month']       = df['datetime'].dt.month
df['year']        = df['datetime'].dt.year
df['day_of_year'] = df['datetime'].dt.dayofyear
df['week_of_year']= df['datetime'].dt.isocalendar().week.astype(int)
df['date_str']    = df['datetime'].dt.strftime('%Y-%m-%d')
df['season']      = df['month'].apply(get_season)

# ── 5.4 Cyclic Features ──
df['hour_sin']  = np.sin(2 * np.pi * df['hour'] / 24)
df['hour_cos']  = np.cos(2 * np.pi * df['hour'] / 24)
df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)
df['dow_sin']   = np.sin(2 * np.pi * df['day_of_week'] / 7)
df['dow_cos']   = np.cos(2 * np.pi * df['day_of_week'] / 7)

# ── 5.5 Flag Features ──
df['is_weekend'] = (df['day_of_week'] >= 5).astype(int)
df['is_holiday'] = df['date_str'].isin(HOLIDAY_SET).astype(int)
df['is_monsoon'] = df['month'].isin([7,8,9]).astype(int)
df['is_summer']  = df['month'].isin([4,5,6]).astype(int)
df['is_winter']  = df['month'].isin([10,11,12,1,2]).astype(int)

# ── 5.6 Season-Aware Peak Flags ──
# Based on EDA: summer/monsoon peaks 12h+22h, winter peaks 10h+18h
def is_morning_peak(row):
    h, s = row['hour'], row['season']
    if s == 'Winter': return int(9 <= h <= 11)
    elif s in ['Summer','Monsoon']: return int(11 <= h <= 13)
    else: return int(10 <= h <= 12)  # Spring

def is_evening_peak(row):
    h, s = row['hour'], row['season']
    if s == 'Winter': return int(17 <= h <= 19)
    elif s in ['Summer','Monsoon']: return int(20 <= h <= 23)
    else: return int(18 <= h <= 21)  # Spring

df['is_morning_peak'] = df.apply(is_morning_peak, axis=1)
df['is_evening_peak'] = df.apply(is_evening_peak, axis=1)

# ── 5.7 Year trend feature (normalized) ──
df['year_norm'] = (df['year'] - 2022) / 4  # 0 to 1 range

print('\n✅ Feature engineering complete')
print(f'Dataset shape: {df.shape}')
```

### 5.8 Aggregate 5-min → Hourly (for training & prediction)

```python
# For prediction we need hourly output (matching Open-Meteo API)
# → Aggregate all 5-min slots into hourly means

df['hour_dt'] = df['datetime'].dt.floor('h')

agg_dict = {r: 'mean' for r in REGIONS}
agg_dict.update({
    'Temperature (°C)': 'mean',
    'Relative Humidity (%)': 'mean',
    'feels_like': 'mean',
    'Precipitation (mm)': 'sum',
    'Wind Speed (m/s)': 'mean',
    'Cloud Cover Total (%)': 'mean',
    'hour': 'first',
    'day_of_week': 'first',
    'month': 'first',
    'year': 'first',
    'day_of_year': 'first',
    'week_of_year': 'first',
    'year_norm': 'first',
    'hour_sin': 'first', 'hour_cos': 'first',
    'month_sin': 'first', 'month_cos': 'first',
    'dow_sin': 'first', 'dow_cos': 'first',
    'is_weekend': 'first',
    'is_holiday': 'first',
    'is_monsoon': 'first',
    'is_summer': 'first',
    'is_winter': 'first',
    'is_morning_peak': 'first',
    'is_evening_peak': 'first',
    'season': 'first',
    'date_str': 'first'
})

df_hourly = df.groupby('hour_dt').agg(agg_dict).reset_index()
df_hourly.rename(columns={'hour_dt':'datetime'}, inplace=True)

print(f'Hourly dataset: {df_hourly.shape}')
print(f'Date range: {df_hourly["datetime"].min()} → {df_hourly["datetime"].max()}')
print(f'Expected rows (4 years + ~4 months): ~{int((df_hourly.shape[0]))} hours')
```

### 5.9 Final Feature List

```python
FEATURES = [
    # Time - Raw
    'hour', 'day_of_week', 'month', 'year', 'day_of_year', 'week_of_year', 'year_norm',
    # Time - Cyclic
    'hour_sin', 'hour_cos', 'month_sin', 'month_cos', 'dow_sin', 'dow_cos',
    # Season-aware peaks
    'is_morning_peak', 'is_evening_peak',
    # Flags
    'is_weekend', 'is_holiday', 'is_monsoon', 'is_summer', 'is_winter',
    # Weather
    'Temperature (°C)', 'Relative Humidity (%)', 'feels_like',
    'Precipitation (mm)', 'Wind Speed (m/s)', 'Cloud Cover Total (%)'
]

print(f'✅ Total features: {len(FEATURES)}')
for i, f in enumerate(FEATURES, 1):
    print(f'  {i:2d}. {f}')

# Verify all features exist
missing = [f for f in FEATURES if f not in df_hourly.columns]
if missing:
    print(f'\n❌ Missing features: {missing}')
else:
    print('\n✅ All features present in dataset')
```

## ✂️ 6. Train / Test Split

```python
# Train: 2022-01-01 to 2025-12-31
# Test : 2026-01-01 to 2026-04-01 (latest available)

train_df = df_hourly[df_hourly['year'] <= 2025].copy()
test_df  = df_hourly[df_hourly['year'] == 2026].copy()

print(f'Train size: {len(train_df):,} hourly records ({train_df["datetime"].min().date()} → {train_df["datetime"].max().date()})')
print(f'Test  size: {len(test_df):,} hourly records ({test_df["datetime"].min().date()} → {test_df["datetime"].max().date()})')
print(f'\nTrain/Test split: {len(train_df)/(len(train_df)+len(test_df))*100:.1f}% / {len(test_df)/(len(train_df)+len(test_df))*100:.1f}%')

X_train = train_df[FEATURES]
X_test  = test_df[FEATURES]

print('\n✅ Split complete')
```

## 🤖 7. Model Architecture & Rationale

### Why these 3 models?

| Model | Why chosen for this task |
|-------|---------------------------|
| **XGBoost** | Proven SOTA for tabular time-series; handles non-linear weather–demand relationships; robust to outliers; fast training |
| **LightGBM** | Gradient boosting with leaf-wise growth → better accuracy on large datasets; handles 26 features efficiently; fastest training; great on cyclic features |
| **CatBoost** | Best handling of categorical-like integer features (hour, month, DOW); symmetric trees reduce variance → prevents overprediction (your earlier variance issue); strong regularization |

### Ensemble Strategy
**Weighted average ensemble** where weights are computed per-region based on each model's MAE on a validation fold.  
Lower MAE → higher weight. This is dynamic — if XGBoost is best for NDMC but LightGBM is best for BRPL, weights adjust accordingly.

## 🏋️ 8. Train Models for All Regions

```python
# ── Helper: Metrics ──
def compute_metrics(y_true, y_pred, region=''):
    mae  = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = np.mean(np.abs((y_true - y_pred) / (y_true + 1e-8))) * 100
    r2   = r2_score(y_true, y_pred)
    acc  = 100 - mape  # accuracy as (100 - MAPE)
    return {'region': region, 'MAE': mae, 'RMSE': rmse, 'MAPE': mape, 'R2': r2, 'Accuracy': acc}

# Storage
all_models    = {}   # {region: {'xgb': model, 'lgb': model, 'cat': model}}
all_weights   = {}   # {region: {'xgb': w, 'lgb': w, 'cat': w}}
all_metrics   = []   # list of dicts
all_preds     = {}   # {region: y_pred_test_ensemble}

# Validation: use last 3 months of training data
val_cutoff = train_df['datetime'].max() - pd.DateOffset(months=3)
train_inner = train_df[train_df['datetime'] <= val_cutoff]
val_df      = train_df[train_df['datetime'] > val_cutoff]

X_train_inner = train_inner[FEATURES]
X_val         = val_df[FEATURES]

print(f'Inner train: {len(train_inner):,} | Validation: {len(val_df):,}')
print('\n' + '='*65)

for region in REGIONS:
    print(f'\n🔄 Training models for: {region}')
    
    y_train_inner = train_inner[region]
    y_val         = val_df[region]
    y_train_full  = train_df[region]
    y_test        = test_df[region]
    
    # ── XGBoost ──
    xgb_model = xgb.XGBRegressor(
        n_estimators      = 1200,
        learning_rate     = 0.04,
        max_depth         = 7,
        subsample         = 0.80,
        colsample_bytree  = 0.80,
        min_child_weight  = 5,
        reg_alpha         = 0.1,
        reg_lambda        = 1.5,
        gamma             = 0.05,
        random_state      = RANDOM_STATE,
        n_jobs            = -1,
        tree_method       = 'hist',
        early_stopping_rounds = 50,
        eval_metric       = 'mae'
    )
    xgb_model.fit(X_train_inner, y_train_inner,
                  eval_set=[(X_val, y_val)], verbose=False)
    xgb_val_pred  = xgb_model.predict(X_val)
    xgb_val_mae   = mean_absolute_error(y_val, xgb_val_pred)
    
    # Retrain on full training data
    xgb_final = xgb.XGBRegressor(
        n_estimators      = xgb_model.best_iteration + 50,
        learning_rate     = 0.04,
        max_depth         = 7,
        subsample         = 0.80,
        colsample_bytree  = 0.80,
        min_child_weight  = 5,
        reg_alpha         = 0.1,
        reg_lambda        = 1.5,
        gamma             = 0.05,
        random_state      = RANDOM_STATE,
        n_jobs            = -1,
        tree_method       = 'hist'
    )
    xgb_final.fit(X_train, y_train_full)
    print(f'  XGBoost  : best_iter={xgb_model.best_iteration}, val_MAE={xgb_val_mae:.2f}')
    
    # ── LightGBM ──
    lgb_model = lgb.LGBMRegressor(
        n_estimators      = 1500,
        learning_rate     = 0.03,
        num_leaves        = 63,
        max_depth         = -1,
        min_child_samples = 20,
        subsample         = 0.80,
        colsample_bytree  = 0.80,
        reg_alpha         = 0.1,
        reg_lambda        = 1.5,
        random_state      = RANDOM_STATE,
        n_jobs            = -1,
        verbose           = -1
    )
    lgb_model.fit(X_train_inner, y_train_inner,
                  eval_set=[(X_val, y_val)],
                  callbacks=[lgb.early_stopping(50, verbose=False), lgb.log_evaluation(-1)])
    lgb_val_pred = lgb_model.predict(X_val)
    lgb_val_mae  = mean_absolute_error(y_val, lgb_val_pred)
    
    lgb_final = lgb.LGBMRegressor(
        n_estimators      = lgb_model.best_iteration_ + 50,
        learning_rate     = 0.03,
        num_leaves        = 63,
        min_child_samples = 20,
        subsample         = 0.80,
        colsample_bytree  = 0.80,
        reg_alpha         = 0.1,
        reg_lambda        = 1.5,
        random_state      = RANDOM_STATE,
        n_jobs            = -1,
        verbose           = -1
    )
    lgb_final.fit(X_train, y_train_full)
    print(f'  LightGBM : best_iter={lgb_model.best_iteration_}, val_MAE={lgb_val_mae:.2f}')
    
    # ── CatBoost ──
    cat_model = CatBoostRegressor(
        iterations         = 1500,
        learning_rate      = 0.04,
        depth              = 7,
        l2_leaf_reg        = 3.0,
        subsample          = 0.80,
        colsample_bylevel  = 0.80,
        min_data_in_leaf   = 20,
        od_type            = 'Iter',
        od_wait            = 50,
        random_seed        = RANDOM_STATE,
        verbose            = 0
    )
    cat_model.fit(X_train_inner, y_train_inner,
                  eval_set=(X_val, y_val))
    cat_val_pred = cat_model.predict(X_val)
    cat_val_mae  = mean_absolute_error(y_val, cat_val_pred)
    
    cat_final = CatBoostRegressor(
        iterations         = cat_model.best_iteration_ + 50,
        learning_rate      = 0.04,
        depth              = 7,
        l2_leaf_reg        = 3.0,
        subsample          = 0.80,
        colsample_bylevel  = 0.80,
        min_data_in_leaf   = 20,
        random_seed        = RANDOM_STATE,
        verbose            = 0
    )
    cat_final.fit(X_train, y_train_full)
    print(f'  CatBoost : best_iter={cat_model.best_iteration_}, val_MAE={cat_val_mae:.2f}')
    
    # ── Compute Weighted Ensemble Weights ──
    # Weight = 1/MAE, normalized. Better model (lower MAE) → higher weight
    inv_maes = np.array([1/xgb_val_mae, 1/lgb_val_mae, 1/cat_val_mae])
    weights  = inv_maes / inv_maes.sum()
    all_weights[region] = {'xgb': weights[0], 'lgb': weights[1], 'cat': weights[2]}
    print(f'  Ensemble weights → XGB:{weights[0]:.3f}, LGB:{weights[1]:.3f}, CAT:{weights[2]:.3f}')
    
    # ── Test Predictions ──
    xgb_test = xgb_final.predict(X_test)
    lgb_test = lgb_final.predict(X_test)
    cat_test = cat_final.predict(X_test)
    ensemble_test = weights[0]*xgb_test + weights[1]*lgb_test + weights[2]*cat_test
    
    all_preds[region] = ensemble_test
    all_models[region] = {'xgb': xgb_final, 'lgb': lgb_final, 'cat': cat_final}
    
    # Metrics
    m_xgb = compute_metrics(y_test, xgb_test, f'{region}_XGB')
    m_lgb = compute_metrics(y_test, lgb_test, f'{region}_LGB')
    m_cat = compute_metrics(y_test, cat_test, f'{region}_CAT')
    m_ens = compute_metrics(y_test, ensemble_test, f'{region}_ENSEMBLE')
    all_metrics.extend([m_xgb, m_lgb, m_cat, m_ens])
    print(f'  ✅ Ensemble Test Accuracy: {m_ens["Accuracy"]:.2f}% | MAPE: {m_ens["MAPE"]:.2f}%')

print('\n' + '='*65)
print('✅ Training complete for all regions!')
```

## 🌤️ 11. Open-Meteo API — Fetch Live Hourly Weather

```python
import openmeteo_requests
import requests_cache
from retry_requests import retry

# Delhi coordinates
DELHI_LAT = 28.6139
DELHI_LON = 77.2090

def fetch_openmeteo_weather(date_str: str):
    """
    Fetch hourly weather for Delhi for a given date from Open-Meteo.
    
    Parameters
    ----------
    date_str : str
        Date in YYYY-MM-DD format, e.g. '2026-05-01'
    
    Returns
    -------
    pd.DataFrame with hourly weather (24 rows)
    """
    cache_session = requests_cache.CachedSession('.cache', expire_after=3600)
    retry_session = retry(cache_session, retries=5, backoff_factor=0.2)
    openmeteo = openmeteo_requests.Client(session=retry_session)
    
    # Determine if forecast or historical
    target_date = pd.to_datetime(date_str)
    today = pd.Timestamp.now().normalize()
    
    if target_date >= today:
        # Forecast API
        url    = 'https://api.open-meteo.com/v1/forecast'
        params = {
            'latitude':  DELHI_LAT,
            'longitude': DELHI_LON,
            'hourly': ['temperature_2m','relative_humidity_2m','apparent_temperature',
                       'precipitation','wind_speed_10m','cloud_cover'],
            'start_date': date_str,
            'end_date':   date_str,
            'timezone':   'Asia/Kolkata'
        }
    else:
        # Historical API
        url    = 'https://archive-api.open-meteo.com/v1/archive'
        params = {
            'latitude':  DELHI_LAT,
            'longitude': DELHI_LON,
            'hourly': ['temperature_2m','relative_humidity_2m','apparent_temperature',
                       'precipitation','wind_speed_10m','cloud_cover'],
            'start_date': date_str,
            'end_date':   date_str,
            'timezone':   'Asia/Kolkata'
        }
    
    responses = openmeteo.weather_api(url, params=params)
    response  = responses[0]
    hourly    = response.Hourly()
    
    hourly_df = pd.DataFrame({
        'datetime':               pd.date_range(
            start   = pd.to_datetime(hourly.Time(), unit='s', utc=True).tz_convert('Asia/Kolkata').tz_localize(None),
            periods = 24, freq='h'
        ),
        'Temperature (°C)':       hourly.Variables(0).ValuesAsNumpy(),
        'Relative Humidity (%)':  hourly.Variables(1).ValuesAsNumpy(),
        'feels_like':             hourly.Variables(2).ValuesAsNumpy(),
        'Precipitation (mm)':     hourly.Variables(3).ValuesAsNumpy(),
        'Wind Speed (m/s)':       hourly.Variables(4).ValuesAsNumpy(),
        'Cloud Cover Total (%)':  hourly.Variables(5).ValuesAsNumpy(),
    })
    
    return hourly_df

print('✅ fetch_openmeteo_weather() function defined')
print('   Supports both forecast (future) and archive (past) dates')
```

## 🔮 12. Prediction Function — Full Pipeline

```python
def build_prediction_features(weather_df: pd.DataFrame, date_str: str) -> pd.DataFrame:
    """
    Build the full feature matrix from hourly weather data for a given date.
    """
    df = weather_df.copy()
    df['datetime']     = pd.to_datetime(df['datetime'])
    
    # Time features
    df['hour']         = df['datetime'].dt.hour
    df['day_of_week']  = df['datetime'].dt.dayofweek
    df['month']        = df['datetime'].dt.month
    df['year']         = df['datetime'].dt.year
    df['day_of_year']  = df['datetime'].dt.dayofyear
    df['week_of_year'] = df['datetime'].dt.isocalendar().week.astype(int)
    df['year_norm']    = (df['year'] - 2022) / 4
    df['season']       = df['month'].apply(get_season)
    
    # Cyclic
    df['hour_sin']  = np.sin(2 * np.pi * df['hour'] / 24)
    df['hour_cos']  = np.cos(2 * np.pi * df['hour'] / 24)
    df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
    df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)
    df['dow_sin']   = np.sin(2 * np.pi * df['day_of_week'] / 7)
    df['dow_cos']   = np.cos(2 * np.pi * df['day_of_week'] / 7)
    
    # Flags
    df['is_weekend']  = (df['day_of_week'] >= 5).astype(int)
    df['is_holiday']  = int(date_str in HOLIDAY_SET)
    df['is_monsoon']  = df['month'].isin([7,8,9]).astype(int)
    df['is_summer']   = df['month'].isin([4,5,6]).astype(int)
    df['is_winter']   = df['month'].isin([10,11,12,1,2]).astype(int)
    
    # Season-aware peaks
    df['is_morning_peak'] = df.apply(is_morning_peak, axis=1)
    df['is_evening_peak'] = df.apply(is_evening_peak, axis=1)
    
    return df[FEATURES]


def predict_demand(date_str: str, models=None, weights=None, verbose=True):
    """
    Predict hourly electricity demand for all Delhi regions for a given date.
    
    Parameters
    ----------
    date_str : str      — Date in YYYY-MM-DD format
    models   : dict     — Loaded models dict (default: all_models)
    weights  : dict     — Loaded weights dict (default: all_weights)
    verbose  : bool     — Print summary
    
    Returns
    -------
    pd.DataFrame with columns: hour, DELHI, BRPL, BYPL, NDPL, NDMC, MES
    """
    if models is None: models = all_models
    if weights is None: weights = all_weights
    
    # 1. Fetch weather
    if verbose: print(f'🌤️  Fetching weather for {date_str}...')
    weather_df = fetch_openmeteo_weather(date_str)
    
    # 2. Build features
    X_pred = build_prediction_features(weather_df, date_str)
    
    # 3. Predict per region
    result_rows = []
    for h in range(24):
        row = {'hour': h, 'time': f'{h:02d}:00'}
        X_h = X_pred.iloc[[h]]
        for r in REGIONS:
            w = weights[r]
            p_xgb = models[r]['xgb'].predict(X_h)[0]
            p_lgb = models[r]['lgb'].predict(X_h)[0]
            p_cat = models[r]['cat'].predict(X_h)[0]
            row[r] = round(w['xgb']*p_xgb + w['lgb']*p_lgb + w['cat']*p_cat, 2)
        result_rows.append(row)
    
    result_df = pd.DataFrame(result_rows)
    
    if verbose:
        date_obj = pd.to_datetime(date_str)
        is_hol = '🎉 HOLIDAY' if date_str in HOLIDAY_SET else ''
        print(f'\n📅 {date_str} ({date_obj.strftime("%A")}) {is_hol}')
        print(result_df[['time']+REGIONS].to_string(index=False))
        print(f'\nPeak Demand (DELHI): {result_df["DELHI"].max():.1f} MW at {result_df.loc[result_df["DELHI"].idxmax(), "time"]}')
        print(f'Min  Demand (DELHI): {result_df["DELHI"].min():.1f} MW at {result_df.loc[result_df["DELHI"].idxmin(), "time"]}')
    
    return result_df


print('✅ predict_demand() function defined')
```

## 📥 14. Load Saved Models & Predict (Deployment Mode)

```python
# ── Run this cell independently to load saved models and predict ──

def load_models_and_predict(date_str: str):
    """
    Load models from disk and predict for a given date.
    Use this in production without needing to retrain.
    """
    import joblib
    
    meta = joblib.load('models/metadata.pkl')
    loaded_models  = {}
    loaded_weights = {}
    
    for r in meta['regions']:
        loaded_models[r] = {
            'xgb': joblib.load(f'models/{r}_xgboost.pkl'),
            'lgb': joblib.load(f'models/{r}_lightgbm.pkl'),
            'cat': joblib.load(f'models/{r}_catboost.pkl')
        }
        loaded_weights[r] = joblib.load(f'models/{r}_weights.pkl')
    
    print('✅ All models loaded from disk')
    return predict_demand(date_str, models=loaded_models, weights=loaded_weights)

# Example usage:
# result = load_models_and_predict('2026-05-15')
print('💡 Usage: result = load_models_and_predict("2026-05-15")')
```
