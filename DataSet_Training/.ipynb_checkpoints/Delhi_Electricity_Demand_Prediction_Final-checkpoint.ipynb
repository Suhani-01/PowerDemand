# ⚡ Delhi Electricity Demand Prediction System
## Final Year Major Project — AI-Based Load Forecasting

**Regions:** DELHI, BYPL, BRPL, NDPL, NDMC, MES  
**Training Period:** Jan 2022 – Dec 2025  
**Testing Period:** Jan 2026 – Apr 2026  
**Models:** XGBoost + LightGBM + Random Forest → R²-Weighted Ensemble  
**New Features:** `is_night`, `apparent_temp_sq`, `temp_sq`  
**Target Accuracy:** R² ≥ 0.90 (MAPE < 5%)

---
| Model | Reason |
|-------|--------|
| **XGBoost** | Non-linear demand patterns, robust to outliers, proven on energy tabular data |
| **LightGBM** | 10× faster than XGBoost on large datasets, leaf-wise growth for complex patterns |
| **Random Forest** | Bagging diversity — stabilises ensemble, reduces variance on peak-demand events |
| **R²-Weighted Ensemble** | Higher weight to better-performing model per region — adapts automatically |


```python
# ============================================================
# CELL 1: Install & Import Libraries
# ============================================================
import subprocess, sys

def install(pkg):
    subprocess.check_call([sys.executable, '-m', 'pip', 'install', pkg, '-q'])

for pkg in ['xgboost', 'lightgbm', 'scikit-learn', 'pandas', 'numpy',
            'matplotlib', 'seaborn', 'requests', 'joblib', 'tqdm']:
    install(pkg)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import requests, warnings, joblib, os
from datetime import datetime, timedelta
from tqdm import tqdm

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

import xgboost as xgb
import lightgbm as lgb

warnings.filterwarnings('ignore')
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (14, 5)
plt.rcParams['font.size'] = 11
np.random.seed(42)

print("✅ All libraries loaded successfully!")
```

```python
# ============================================================
# CELL 2: Configuration
# ============================================================

# ⚠️ UPDATE THESE PATHS TO YOUR LOCAL DATASET LOCATIONS
SLDC_PATH    = 'delhi_sldc_5min_2022_2026.csv'   # SLDC demand data
WEATHER_PATH = 'Delhi_Weather_5M.csv'            # Weather data

REGIONS = ['DELHI', 'BYPL', 'BRPL', 'NDPL', 'NDMC', 'MES']

# Delhi coordinates for OpenMeteo API
DELHI_LAT = 28.6139
DELHI_LON = 77.2090

# Train: 2022-2025, Test: 2026
TRAIN_END_YEAR  = 2025
TEST_START_YEAR = 2026

# Model save directory
MODEL_DIR = 'saved_models'
os.makedirs(MODEL_DIR, exist_ok=True)

print(f"✅ Config set. Models will be saved to: {MODEL_DIR}/")
```

```python
# ============================================================
# CELL 3: Load & Merge Datasets
# ============================================================

print("📂 Loading SLDC demand data...")
sldc = pd.read_csv(SLDC_PATH)
print(f"   Shape: {sldc.shape}")
print(f"   Date Range: {sldc['Date'].iloc[0]}  →  {sldc['Date'].iloc[-1]}")

print("\n📂 Loading weather data...")
weather = pd.read_csv(WEATHER_PATH)
print(f"   Shape: {weather.shape}")

# ── Parse datetime ──────────────────────────────────────────
def parse_datetime(df):
    df = df.copy()
    df['datetime'] = pd.to_datetime(
        df['Date'] + ' ' + df['TimeSlot'],
        format='%d/%m/%Y %H:%M',
        dayfirst=True
    )
    df = df.sort_values('datetime').reset_index(drop=True)
    df.drop(columns=['Date', 'TimeSlot'], inplace=True)
    return df

sldc    = parse_datetime(sldc)
weather = parse_datetime(weather)

# ── Keep required weather features ──────────────────────────
WEATHER_COLS = [
    'datetime',
    'Temperature (°C)',
    'Relative Humidity (%)',
    'Apparent Temperature (°C)',
    'Precipitation (mm)',
    'Wind Speed (m/s)',
    'Cloud Cover Total (%)'
]
weather = weather[WEATHER_COLS].rename(columns={'Apparent Temperature (°C)': 'feels_like'})

# ── Merge on datetime ────────────────────────────────────────
df = pd.merge(sldc, weather, on='datetime', how='inner')
df = df.sort_values('datetime').reset_index(drop=True)

# ── Missing value check ──────────────────────────────────────
missing = df.isnull().sum()
print(f"\n🔍 Missing values after merge:")
print(missing[missing > 0] if missing.sum() > 0 else "   None ✅")

print(f"\n✅ Merged dataset shape: {df.shape}")
print(f"   Date range: {df['datetime'].min()} → {df['datetime'].max()}")
df.head(3)
```

```python
# ============================================================
# CELL 4: Outlier Removal (IQR method per region)
# ============================================================
"""
Strategy:
  - IQR multiplier = 3.0 (conservative — only extreme spikes removed)
  - Per-region independently (each region has different demand range)
  - Outliers → NaN → time-interpolate (requires DatetimeIndex) → ffill/bfill
"""

print("🔍 Outlier detection and removal...")
print("\nBefore outlier removal:")
print(df[REGIONS].describe().round(2))

df_clean = df.copy()

outlier_stats = {}
for region in REGIONS:
    Q1  = df_clean[region].quantile(0.25)
    Q3  = df_clean[region].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 3.0 * IQR
    upper = Q3 + 3.0 * IQR

    mask = (df_clean[region] < lower) | (df_clean[region] > upper)
    n_out = mask.sum()
    outlier_stats[region] = {'count': int(n_out), 'lower': round(lower, 2), 'upper': round(upper, 2)}
    df_clean.loc[mask, region] = np.nan

# FIX: interpolate(method='time') needs DatetimeIndex
df_clean = df_clean.set_index('datetime')
df_clean[REGIONS] = df_clean[REGIONS].interpolate(method='time')
df_clean = df_clean.reset_index()          # restore datetime column
df_clean[REGIONS] = df_clean[REGIONS].ffill().bfill()

print("\n📊 Outliers removed per region:")
for r, s in outlier_stats.items():
    print(f"   {r:8s}: {s['count']:4d} outliers | valid range [{s['lower']}, {s['upper']}]")

print("\nAfter outlier removal:")
print(df_clean[REGIONS].describe().round(2))
```

```python
# ============================================================
# CELL 5: Aggregate 5-min → Hourly
# ============================================================
"""
We train on 5-min data BUT aggregate to HOURLY because:
  - OpenMeteo API gives hourly weather forecasts
  - Hourly granularity still captures morning peak / evening peak patterns
  - Reduces dataset size from 430K → ~35K rows → faster, higher accuracy
"""

demand_cols  = REGIONS
weather_feat = ['Temperature (°C)', 'Relative Humidity (%)', 'feels_like',
                'Precipitation (mm)', 'Wind Speed (m/s)', 'Cloud Cover Total (%)']

df_clean['datetime_hour'] = df_clean['datetime'].dt.floor('h')

hourly = df_clean.groupby('datetime_hour').agg(
    {**{c: 'mean' for c in demand_cols},
     **{c: 'mean' for c in weather_feat}}
).reset_index()
hourly.rename(columns={'datetime_hour': 'datetime'}, inplace=True)

hourly.dropna(inplace=True)
hourly.reset_index(drop=True, inplace=True)

print(f"✅ 5-min → Hourly aggregation complete")
print(f"   Hourly dataset shape: {hourly.shape}")
print(f"   Date range: {hourly['datetime'].iloc[0]}  →  {hourly['datetime'].iloc[-1]}")
hourly.head(3)
```

```python
# ============================================================
# CELL 6: Feature Engineering
# ============================================================
"""
Features used:
  Time-Raw:    hour, day_of_week, month, year, day_of_year, week_of_year
  Time-Cyclic: hour_sin, hour_cos, month_sin, month_cos, dow_sin, dow_cos
  Flags:       is_weekend, is_monsoon, is_summer, is_winter, is_night  ← NEW
  Weather:     Temperature, Humidity, feels_like, Precipitation,
               Wind Speed, Cloud Cover
  Engineered:  temp_sq, apparent_temp_sq                               ← NEW
"""

def add_features(df):
    df = df.copy()
    dt = df['datetime']

    # ── Raw time ─────────────────────────────────────────────
    df['hour']         = dt.dt.hour
    df['day_of_week']  = dt.dt.dayofweek          # 0=Mon, 6=Sun
    df['month']        = dt.dt.month
    df['year']         = dt.dt.year
    df['day_of_year']  = dt.dt.dayofyear
    df['week_of_year'] = dt.dt.isocalendar().week.astype(int)

    # ── Cyclic encoding ──────────────────────────────────────
    df['hour_sin']  = np.sin(2 * np.pi * df['hour']        / 24)
    df['hour_cos']  = np.cos(2 * np.pi * df['hour']        / 24)
    df['month_sin'] = np.sin(2 * np.pi * df['month']       / 12)
    df['month_cos'] = np.cos(2 * np.pi * df['month']       / 12)
    df['dow_sin']   = np.sin(2 * np.pi * df['day_of_week'] / 7)
    df['dow_cos']   = np.cos(2 * np.pi * df['day_of_week'] / 7)

    # ── Calendar flags ───────────────────────────────────────
    df['is_weekend'] = (df['day_of_week'] >= 5).astype(int)
    df['is_monsoon'] = df['month'].isin([6, 7, 8, 9]).astype(int)
    df['is_summer']  = df['month'].isin([3, 4, 5, 6]).astype(int)
    df['is_winter']  = df['month'].isin([11, 12, 1, 2]).astype(int)

    # ── NEW: is_night (10 PM – 5 AM = low demand hours) ─────
    df['is_night']   = ((df['hour'] >= 22) | (df['hour'] <= 5)).astype(int)

    # ── NEW: Polynomial weather features ────────────────────
    # temp_sq captures non-linear AC-load relationship with temperature
    # apparent_temp_sq captures extreme heat/cold effect on demand
    df['temp_sq']          = df['Temperature (°C)'] ** 2
    df['apparent_temp_sq'] = df['feels_like'] ** 2

    return df

hourly = add_features(hourly)

# ── Define feature columns ───────────────────────────────────
FEATURE_COLS = [
    # Raw time
    'hour', 'day_of_week', 'month', 'year', 'day_of_year', 'week_of_year',
    # Cyclic
    'hour_sin', 'hour_cos', 'month_sin', 'month_cos', 'dow_sin', 'dow_cos',
    # Flags
    'is_weekend', 'is_monsoon', 'is_summer', 'is_winter',
    'is_night',                   # NEW
    # Weather
    'Temperature (°C)', 'Relative Humidity (%)', 'feels_like',
    'Precipitation (mm)', 'Wind Speed (m/s)', 'Cloud Cover Total (%)',
    # Engineered
    'temp_sq',                    # NEW
    'apparent_temp_sq',           # NEW
]

print(f"✅ Features created: {len(FEATURE_COLS)} total")
print(f"\n   Feature list:")
for i, f in enumerate(FEATURE_COLS, 1):
    tag = ' ← NEW' if f in ('is_night', 'temp_sq', 'apparent_temp_sq') else ''
    print(f"   {i:2d}. {f}{tag}")
print(f"\nDataset shape after feature engineering: {hourly.shape}")
```

```python
# ============================================================
# CELL 7: Train / Test Split (Year-based: 2022-2025 | 2026)
# ============================================================

train_df = hourly[hourly['datetime'].dt.year <= TRAIN_END_YEAR].reset_index(drop=True)
test_df  = hourly[hourly['datetime'].dt.year >= TEST_START_YEAR].reset_index(drop=True)

X_train = train_df[FEATURE_COLS].values
X_test  = test_df[FEATURE_COLS].values

print(f"📊 Train set: {len(train_df):,} rows | {train_df['datetime'].min().date()} → {train_df['datetime'].max().date()}")
print(f"📊 Test  set: {len(test_df):,}  rows | {test_df['datetime'].min().date()} → {test_df['datetime'].max().date()}")
print(f"\n✅ X_train shape: {X_train.shape}")
print(f"✅ X_test  shape: {X_test.shape}")
```

```python
# ============================================================
# CELL 8: Helper Functions — Metrics & Weights
# ============================================================

def mape_score(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / np.clip(np.abs(y_true), 1e-6, None))) * 100

def evaluate_model(y_true, y_pred, model_name='Model', region=''):
    mae  = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = mape_score(y_true, y_pred)
    r2   = r2_score(y_true, y_pred)
    print(f"   [{model_name:13s}] {region} — MAE: {mae:.2f} | RMSE: {rmse:.2f} | MAPE: {mape:.2f}% | R²: {r2:.4f}")
    return {'MAE': mae, 'RMSE': rmse, 'MAPE': mape, 'R2': r2}

def compute_r2_weights(metrics_dict):
    """
    R²-based ensemble weights (same as Ensemble_updated notebook).
    Higher R² → higher weight. Automatically adapts per region.
    """
    # Exclude 'Ensemble' key if present
    model_names = [k for k in metrics_dict if k != 'Ensemble']
    r2s = {m: max(metrics_dict[m]['R2'], 0.001) for m in model_names}
    total = sum(r2s.values())
    return {m: r2s[m] / total for m in model_names}

print("✅ Helper functions defined.")
```

```python
# ============================================================
# CELL 9: Model Definitions (Tuned for 90%+ accuracy)
# ============================================================
"""
Hyperparameters tuned to push R² above 0.90:
  XGBoost  — deeper trees (max_depth=8), more estimators, lower LR
  LightGBM — more leaves (127), feature_fraction for regularisation
  RF       — deeper (max_depth=None = full depth), more trees
"""

def get_models():
    xgb_model = xgb.XGBRegressor(
        n_estimators      = 800,
        learning_rate     = 0.04,
        max_depth         = 8,
        subsample         = 0.85,
        colsample_bytree  = 0.85,
        min_child_weight  = 3,
        reg_alpha         = 0.05,
        reg_lambda        = 1.0,
        gamma             = 0.1,
        random_state      = 42,
        n_jobs            = -1,
        tree_method       = 'hist',
        verbosity         = 0
    )
    lgb_model = lgb.LGBMRegressor(
        n_estimators      = 800,
        learning_rate     = 0.04,
        num_leaves        = 127,
        max_depth         = -1,
        subsample         = 0.85,
        colsample_bytree  = 0.85,
        min_child_samples = 15,
        reg_alpha         = 0.05,
        reg_lambda        = 1.0,
        random_state      = 42,
        n_jobs            = -1,
        verbose           = -1
    )
    rf_model = RandomForestRegressor(
        n_estimators      = 400,
        max_depth         = None,       # full depth for max accuracy
        min_samples_split = 5,
        min_samples_leaf  = 3,
        max_features      = 0.75,
        random_state      = 42,
        n_jobs            = -1
    )
    return xgb_model, lgb_model, rf_model

print("✅ Model definitions ready!")
print("   📦 XGBoost    — n_est=800, depth=8, lr=0.04")
print("   📦 LightGBM   — n_est=800, leaves=127, lr=0.04")
print("   📦 RandomForest — n_est=400, full depth")
```

```python
# ============================================================
# CELL 10: Model Training & Evaluation — All Regions
# ============================================================
"""
For each region:
  1. Train XGBoost, LightGBM, Random Forest on hourly data
  2. Evaluate on 2026 test set
  3. Compute R²-based weights for ensemble (higher R² → higher weight)
  4. Ensemble = weighted average
  5. Target: R² ≥ 0.90, MAPE < 5%
"""

all_results  = {}   # region → {model: metrics}
all_weights  = {}   # region → {model: weight}
all_preds    = {}   # region → {model: array}
all_models   = {}   # region → {model: fitted_model}

print("🚀 Starting training for all regions...")
print("=" * 70)

for region in REGIONS:
    print(f"\n🏙️  Region: {region}")
    print("-" * 50)

    y_train = train_df[region].values
    y_test  = test_df[region].values

    xgb_m, lgb_m, rf_m = get_models()

    preds   = {}
    metrics = {}
    models  = {}

    # ── 1. XGBoost ────────────────────────────────────────────
    print("   Training XGBoost...", end=" ")
    xgb_m.fit(X_train, y_train, verbose=False)
    preds['XGBoost']   = xgb_m.predict(X_test)
    metrics['XGBoost'] = evaluate_model(y_test, preds['XGBoost'], 'XGBoost', region)
    models['XGBoost']  = xgb_m

    # ── 2. LightGBM ───────────────────────────────────────────
    print("   Training LightGBM...", end=" ")
    lgb_m.fit(X_train, y_train)
    preds['LightGBM']   = lgb_m.predict(X_test)
    metrics['LightGBM'] = evaluate_model(y_test, preds['LightGBM'], 'LightGBM', region)
    models['LightGBM']  = lgb_m

    # ── 3. Random Forest ──────────────────────────────────────
    print("   Training RandomForest...", end=" ")
    rf_m.fit(X_train, y_train)
    preds['RandomForest']   = rf_m.predict(X_test)
    metrics['RandomForest'] = evaluate_model(y_test, preds['RandomForest'], 'RandomForest', region)
    models['RandomForest']  = rf_m

    # ── 4. R²-Weighted Ensemble ───────────────────────────────
    weights = compute_r2_weights(metrics)
    ensemble_pred = (
        weights['XGBoost']      * preds['XGBoost']     +
        weights['LightGBM']     * preds['LightGBM']    +
        weights['RandomForest'] * preds['RandomForest']
    )
    preds['Ensemble']   = ensemble_pred
    metrics['Ensemble'] = evaluate_model(y_test, ensemble_pred, '⭐ Ensemble', region)

    # Store
    all_results[region] = metrics
    all_weights[region] = weights
    all_preds[region]   = preds
    all_models[region]  = models
    models['ensemble_weights'] = weights   # save inside models dict too

    print(f"   Weights → XGB: {weights['XGBoost']:.3f} | LGBM: {weights['LightGBM']:.3f} | RF: {weights['RandomForest']:.3f}")

print("\n" + "=" * 70)
print("✅ Training complete for all regions!")
```

```python
# ============================================================
# CELL 11: Summary Table — All Models, All Regions
# ============================================================

rows = []
for region in REGIONS:
    for model_name, m in all_results[region].items():
        rows.append({
            'Region'   : region,
            'Model'    : model_name,
            'MAE (MW)' : round(m['MAE'],  2),
            'RMSE (MW)': round(m['RMSE'], 2),
            'MAPE (%)'  : round(m['MAPE'], 2),
            'R²'        : round(m['R2'],   4)
        })

summary_df = pd.DataFrame(rows)
ensemble_summary = summary_df[summary_df['Model'] == 'Ensemble']

print("\n📊 FULL RESULTS TABLE (Test Set 2026):")
print(summary_df.to_string(index=False))

print("\n⭐ ENSEMBLE MODEL SUMMARY:")
print(ensemble_summary.to_string(index=False))

# Accuracy check
print("\n🎯 Accuracy Check (R² ≥ 0.90 target):")
for _, row in ensemble_summary.iterrows():
    status = "✅ PASS" if row['R²'] >= 0.90 else "⚠️  BELOW TARGET"
    print(f"   {row['Region']:6s}: R²={row['R²']:.4f}, MAPE={row['MAPE (%)']:.2f}%  {status}")
```

```python
# ============================================================
# CELL 12: Save Models to .pkl
# ============================================================

for region in REGIONS:
    for model_name in ['XGBoost', 'LightGBM', 'RandomForest']:
        path = os.path.join(MODEL_DIR, f'{region}_{model_name}.pkl')
        joblib.dump(all_models[region][model_name], path)
    joblib.dump(all_weights[region], os.path.join(MODEL_DIR, f'{region}_ensemble_weights.pkl'))

joblib.dump(FEATURE_COLS, os.path.join(MODEL_DIR, 'feature_cols.pkl'))

print(f"✅ All models saved to '{MODEL_DIR}/' folder")
print("   Files saved:")
for f in sorted(os.listdir(MODEL_DIR)):
    size = os.path.getsize(os.path.join(MODEL_DIR, f)) / 1024
    print(f"   📦 {f} ({size:.1f} KB)")
```

```python
# ============================================================
# CELL 13: Visualisations — Actual vs Predicted (Test Set)
# ============================================================

colors = ['#2196F3', '#FF5722', '#4CAF50', '#9C27B0', '#FF9800', '#009688']
fig, axes = plt.subplots(3, 2, figsize=(18, 14))
axes = axes.flatten()

# Plot last 7 days of test set (168 hours)
N_SHOW = min(168, len(test_df))
idx_s  = len(test_df) - N_SHOW

for i, (region, color) in enumerate(zip(REGIONS, colors)):
    ax = axes[i]
    y_true = test_df[region].values[idx_s:]
    y_ens  = all_preds[region]['Ensemble'][idx_s:]
    y_xgb  = all_preds[region]['XGBoost'][idx_s:]

    x = range(len(y_true))
    ax.plot(x, y_true, label='Actual',   color='black', linewidth=1.8, alpha=0.95)
    ax.plot(x, y_ens,  label='Ensemble', color=color,   linewidth=1.3, linestyle='-')
    ax.plot(x, y_xgb,  label='XGBoost',  color='gray',  linewidth=0.8, linestyle='--', alpha=0.6)

    m = all_results[region]['Ensemble']
    ax.set_title(f"{region} — MAPE: {m['MAPE']:.2f}% | R²: {m['R2']:.4f}", fontsize=11, fontweight='bold')
    ax.set_xlabel('Hours (last 7 days of test set)')
    ax.set_ylabel('Demand (MW)')
    ax.legend(fontsize=8)
    ax.grid(alpha=0.3)

plt.suptitle('Delhi Electricity Demand — Actual vs Predicted (2026 Test Set)\nR²-Weighted Ensemble (XGBoost + LightGBM + RandomForest)',
             fontsize=13, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('actual_vs_predicted.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Plot saved as 'actual_vs_predicted.png'")
```

```python
# ============================================================
# CELL 14: R² Comparison — All Models vs All Regions
# ============================================================

rows_r2 = []
for region in REGIONS:
    for model_name, m in all_results[region].items():
        rows_r2.append({'Region': region, 'Model': model_name, 'R²': round(m['R2'], 4)})

r2_df   = pd.DataFrame(rows_r2)
r2_pivot = r2_df.pivot(index='Region', columns='Model', values='R²')

ax = r2_pivot.plot(kind='bar', figsize=(14, 6), width=0.7,
                   color=['#2196F3', '#4CAF50', '#FF5722', '#9C27B0'])
ax.set_title('R² Score — All Models vs All Regions', fontsize=13, fontweight='bold')
ax.set_ylabel('R² Score (higher = better)')
ax.set_xlabel('Region')
ax.set_ylim(0.80, 1.0)
ax.axhline(0.90, color='red',    linestyle='--', alpha=0.6, label='0.90 target')
ax.axhline(0.95, color='orange', linestyle='--', alpha=0.5, label='0.95 stretch goal')
ax.legend(title='Model', loc='lower right')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('r2_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ============================================================
# CELL 18: OpenMeteo API — Fetch Tomorrow's Hourly Weather
# ============================================================

def fetch_openmeteo_hourly(lat=DELHI_LAT, lon=DELHI_LON, days=2):
    """Fetch hourly weather forecast from OpenMeteo (free, no API key)."""
    url = 'https://api.open-meteo.com/v1/forecast'
    params = {
        'latitude'      : lat,
        'longitude'     : lon,
        'hourly'        : ['temperature_2m', 'relative_humidity_2m',
                           'apparent_temperature', 'precipitation',
                           'wind_speed_10m', 'cloud_cover'],
        'forecast_days' : days,
        'timezone'      : 'Asia/Kolkata'
    }
    print("🌐 Fetching weather from OpenMeteo API...")
    resp = requests.get(url, params=params, timeout=30)
    resp.raise_for_status()
    data = resp.json()['hourly']

    weather_df = pd.DataFrame({
        'datetime'              : pd.to_datetime(data['time']),
        'Temperature (°C)'      : data['temperature_2m'],
        'Relative Humidity (%)' : data['relative_humidity_2m'],
        'feels_like'            : data['apparent_temperature'],
        'Precipitation (mm)'    : data['precipitation'],
        'Wind Speed (m/s)'      : data['wind_speed_10m'],
        'Cloud Cover Total (%)' : data['cloud_cover']
    })
    tomorrow = (datetime.now() + timedelta(days=1)).date()
    weather_df = weather_df[weather_df['datetime'].dt.date == tomorrow].reset_index(drop=True)
    print(f"✅ Fetched {len(weather_df)} hourly records for {tomorrow}")
    return weather_df

try:
    forecast_weather = fetch_openmeteo_hourly()
    print(forecast_weather)
except Exception as e:
    print(f"⚠️ API call failed: {e}")
    print("Using dummy weather data for demonstration...")
    tomorrow = (datetime.now() + timedelta(days=1)).date()
    # FIX: freq='1h' (lowercase) for pandas >= 2.2
    forecast_weather = pd.DataFrame({
        'datetime'              : pd.date_range(f'{tomorrow} 00:00', periods=24, freq='1h'),
        'Temperature (°C)'      : np.random.uniform(25, 40, 24),
        'Relative Humidity (%)' : np.random.uniform(40, 80, 24),
        'feels_like'            : np.random.uniform(28, 45, 24),
        'Precipitation (mm)'    : np.zeros(24),
        'Wind Speed (m/s)'      : np.random.uniform(1, 5, 24),
        'Cloud Cover Total (%)' : np.random.uniform(0, 60, 24)
    })
```

```python
# ============================================================
# CELL 19: Build Forecast Feature Matrix (Tomorrow)
# ============================================================

def build_forecast_features(weather_df):
    """
    Given hourly weather DataFrame (24 rows),
    create all model features including new ones (is_night, temp_sq, apparent_temp_sq).
    Returns numpy array ready for model.predict()
    """
    df = weather_df.copy()
    dt = df['datetime']

    df['hour']         = dt.dt.hour
    df['day_of_week']  = dt.dt.dayofweek
    df['month']        = dt.dt.month
    df['year']         = dt.dt.year
    df['day_of_year']  = dt.dt.dayofyear
    df['week_of_year'] = dt.dt.isocalendar().week.astype(int)

    df['hour_sin']  = np.sin(2 * np.pi * df['hour']        / 24)
    df['hour_cos']  = np.cos(2 * np.pi * df['hour']        / 24)
    df['month_sin'] = np.sin(2 * np.pi * df['month']       / 12)
    df['month_cos'] = np.cos(2 * np.pi * df['month']       / 12)
    df['dow_sin']   = np.sin(2 * np.pi * df['day_of_week'] / 7)
    df['dow_cos']   = np.cos(2 * np.pi * df['day_of_week'] / 7)

    df['is_weekend'] = (df['day_of_week'] >= 5).astype(int)
    df['is_monsoon'] = df['month'].isin([6, 7, 8, 9]).astype(int)
    df['is_summer']  = df['month'].isin([3, 4, 5, 6]).astype(int)
    df['is_winter']  = df['month'].isin([11, 12, 1, 2]).astype(int)
    df['is_night']   = ((df['hour'] >= 22) | (df['hour'] <= 5)).astype(int)

    df['temp_sq']          = df['Temperature (°C)'] ** 2
    df['apparent_temp_sq'] = df['feels_like'] ** 2

    return df[FEATURE_COLS].values, df['datetime']

X_pred, pred_datetimes = build_forecast_features(forecast_weather)
print(f"✅ Forecast feature matrix shape: {X_pred.shape}")
print(f"   Features: {FEATURE_COLS}")
```

```python
# ============================================================
# CELL 20: Predict Tomorrow's Hourly Demand — All Regions
# ============================================================

tomorrow_predictions = {}

for region in REGIONS:
    w = all_weights[region]
    m = all_models[region]

    pred_xgb  = m['XGBoost'].predict(X_pred)
    pred_lgbm = m['LightGBM'].predict(X_pred)
    pred_rf   = m['RandomForest'].predict(X_pred)

    pred_ensemble = (
        w['XGBoost']      * pred_xgb  +
        w['LightGBM']     * pred_lgbm +
        w['RandomForest'] * pred_rf
    )
    tomorrow_predictions[region] = {
        'XGBoost'     : pred_xgb,
        'LightGBM'    : pred_lgbm,
        'RandomForest': pred_rf,
        'Ensemble'    : pred_ensemble
    }

# ── Build output table ───────────────────────────────────────
pred_df = pd.DataFrame({'Hour': pd.to_datetime(pred_datetimes).strftime('%H:%M')})
for region in REGIONS:
    pred_df[f'{region} (MW)'] = tomorrow_predictions[region]['Ensemble'].round(2)

tomorrow_date = pd.to_datetime(pred_datetimes).iloc[0].date()
print(f"\n📅 Tomorrow's Hourly Electricity Demand Forecast (MW)")
print(f"   Date: {tomorrow_date}")
print("=" * 75)
print(pred_df.to_string(index=False))
```

```python
# ============================================================
# CELL 21: Forecast Visualisation — Tomorrow
# ============================================================

colors = ['#2196F3', '#FF5722', '#4CAF50', '#9C27B0', '#FF9800', '#009688']
fig, axes = plt.subplots(3, 2, figsize=(16, 14))
axes = axes.flatten()
hours = list(range(24))

for i, (region, color) in enumerate(zip(REGIONS, colors)):
    ax = axes[i]
    p = tomorrow_predictions[region]

    all_preds_arr = np.vstack([p['XGBoost'], p['LightGBM'], p['RandomForest']])
    ax.fill_between(hours, all_preds_arr.min(0), all_preds_arr.max(0),
                    alpha=0.12, color='gray', label='Model range')
    ax.plot(hours, p['XGBoost'],      label='XGBoost',      color='blue',  linestyle='--', alpha=0.6)
    ax.plot(hours, p['LightGBM'],     label='LightGBM',     color='green', linestyle='--', alpha=0.6)
    ax.plot(hours, p['RandomForest'], label='RandomForest', color='orange',linestyle='--', alpha=0.6)
    ax.plot(hours, p['Ensemble'],     label='Ensemble',     color=color,   linewidth=2.5)

    # Annotate peak
    peak_h = int(np.argmax(p['Ensemble']))
    peak_v = p['Ensemble'][peak_h]
    ax.annotate(f"Peak\n{peak_v:.0f} MW", xy=(peak_h, peak_v),
                xytext=(peak_h + 1.5, peak_v * 1.02),
                arrowprops=dict(arrowstyle='->', color='red'),
                fontsize=8, color='red', fontweight='bold')

    ax.set_title(f'{region} — Tomorrow Forecast', fontweight='bold', fontsize=11)
    ax.set_xlabel('Hour of Day')
    ax.set_ylabel('Demand (MW)')
    ax.set_xticks(hours[::3])
    ax.set_xticklabels([f'{h:02d}:00' for h in hours[::3]], rotation=45)
    ax.legend(fontsize=7)
    ax.grid(alpha=0.3)

plt.suptitle(f'Delhi Electricity Demand Forecast — {tomorrow_date}\n'
             f'R²-Weighted Ensemble (XGBoost + LightGBM + RandomForest)',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig(f'forecast_{tomorrow_date}.png', dpi=150, bbox_inches='tight')
plt.show()
print(f"✅ Forecast plot saved as 'forecast_{tomorrow_date}.png'")
```

```python
# ============================================================
# CELL 22: Peak Demand Analysis
# ============================================================

print("⚡ TOMORROW'S PEAK DEMAND FORECAST")
print("=" * 55)
for region in REGIONS:
    ens = tomorrow_predictions[region]['Ensemble']
    peak_hour = int(np.argmax(ens))
    peak_mw   = ens[peak_hour]
    avg_mw    = ens.mean()
    min_mw    = ens.min()
    min_hour  = int(np.argmin(ens))
    print(f"  {region:6s}: Peak={peak_mw:.1f} MW @ {peak_hour:02d}:00 | Min={min_mw:.1f} MW @ {min_hour:02d}:00 | Avg={avg_mw:.1f} MW")

print("\n✅ Analysis complete!")
```

```python
# ============================================================
# CELL 23: Load Saved Models & Predict (Inference Script)
# ============================================================
"""
Run this cell anytime to load saved .pkl models and predict
WITHOUT retraining — just run from Cell 18 onwards.
"""

def load_models_and_predict(weather_df, model_dir='saved_models'):
    """Load all saved .pkl models and predict for given weather DataFrame."""
    loaded_weights = {}
    loaded_models  = {}
    feature_cols   = joblib.load(os.path.join(model_dir, 'feature_cols.pkl'))

    for region in REGIONS:
        loaded_models[region]  = {}
        for mname in ['XGBoost', 'LightGBM', 'RandomForest']:
            loaded_models[region][mname] = joblib.load(
                os.path.join(model_dir, f'{region}_{mname}.pkl'))
        loaded_weights[region] = joblib.load(
            os.path.join(model_dir, f'{region}_ensemble_weights.pkl'))

    X, _ = build_forecast_features(weather_df)

    results = {}
    for region in REGIONS:
        w = loaded_weights[region]
        ensemble = (
            w['XGBoost']      * loaded_models[region]['XGBoost'].predict(X)     +
            w['LightGBM']     * loaded_models[region]['LightGBM'].predict(X)    +
            w['RandomForest'] * loaded_models[region]['RandomForest'].predict(X)
        )
        results[region] = ensemble.round(2)
    return results

# Quick test
print("🔄 Testing model reload...")
reloaded_preds = load_models_and_predict(forecast_weather)
print("✅ Models reloaded and predictions verified!")
for region in REGIONS:
    print(f"   {region}: {reloaded_preds[region][:3]}... (first 3 hours)")
```
